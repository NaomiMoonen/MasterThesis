```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import squarify
```

Change the directory and read in the deald data that was obtained from the ThomsonOne database.

```python
def find(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)
```

```python
path = find('dealdata.xlsx', 'C://')
data = pd.read_excel(path)
```

Only keep deals that have an effective date before 2018

```python
data_new = data[data['Date Effective']<'2018-01-01']
```

```python
data = data_new.copy()
```

```python
data.shape
```

Sort the deals by deal value

```python
data.sort_values(by='Value of Transaction ($mil)', ascending=False)
```

Plot the number of deals per year of their effective date.

```python
#Deals per date
plt.figure(figsize=(10,6))
years = data['Date Effective'].dt.year
plt.xticks(fontname="century", fontsize=16)
plt.yticks(fontname="century", fontsize=16)
n, bins, patches =plt.hist(years,bins=[2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018], align='left', color="gray")
```

Plot the industries of the targets and acquirers.

```python
#Industries of the targets
plt.figure(figsize=(12,5))
plt.xticks(fontname="century", fontsize=14)
plt.yticks(fontname="century", fontsize=14)
ax = data['Target Macro Industry'].value_counts(ascending=True).plot('barh', color="gray")
for i in ax.patches:
    ax.text(i.get_width()+5, i.get_y(), str(i.get_width()), fontsize=14, fontname="century")
    ax.plot(i.get_width()+5, i.get_y(),color="black")
plt.show()

#Industries of the acquirors
plt.figure(figsize=(12,5))
plt.xticks(fontname="century", fontsize=14)
plt.yticks(fontname="century", fontsize=14)
ax = data['Acquiror Macro Industry'].value_counts(ascending=True).plot('barh', color="gray")
for i in ax.patches:
    ax.text(i.get_width()+4, i.get_y(), str(i.get_width()), fontsize=14, fontname="century")
plt.show()
```

Compare deals with industries financials and healthcare by counting how often these are the only two industries and in which combinations.

```python
fin_hc = data[((data['Target Macro Industry']=='Financials') & (data['Acquiror Macro Industry']=='Financials'))|
             ((data['Target Macro Industry']=='Financials')&(data['Acquiror Macro Industry']=='Healthcare'))|
             ((data['Target Macro Industry']=='Healthcare')&(data['Acquiror Macro Industry']=='Financials'))|
             ((data['Target Macro Industry']=='Healthcare')&(data['Acquiror Macro Industry']=='Healthcare'))]
fin = data[((data['Target Macro Industry']=='Financials') & (data['Acquiror Macro Industry']=='Financials'))]
print(fin.shape)
finhc = data[((data['Target Macro Industry']=='Financials')&(data['Acquiror Macro Industry']=='Healthcare'))]
print(finhc.shape)
hcfin = data[((data['Target Macro Industry']=='Healthcare')&(data['Acquiror Macro Industry']=='Financials'))]
print(hcfin.shape)
hc = data[((data['Target Macro Industry']=='Healthcare')&(data['Acquiror Macro Industry']=='Healthcare'))]
print(hc.shape)
print(fin_hc.shape)
```

Plot these counts of healthcare and financial deals as a treemap, where the size of the rectangle represents the count.

```python
plt.figure(figsize=(20,10))
squarify.plot(sizes=[168, 57, 159, 211], label=["Both financials (168)", "Target Financials, \n Acquirer Healthcare (57)", "Target Healthcare, \n Acquirer Financials (159)", "Both Healthcare (211)"], alpha=.7,
             color=["#000000","#808080", "#A0A0A0", '#505050'],text_kwargs={'fontsize':30,'fontname':'century','color':'white'})
plt.axis('off')
plt.show()
```

Plot the target and acquirer nations.

```python
#Target nation
plt.figure(figsize=(14,5))
plt.xticks(fontname="century", fontsize=14)
plt.yticks(fontname="century", fontsize=14)
ax = data['Target Nation'].value_counts(ascending=False)[0:10].sort_values().plot('barh',color="gray")
for i in ax.patches:
    ax.text(i.get_width()+5, i.get_y(), str(i.get_width()), fontsize=14, fontname="century")
plt.show()

#Acquiror nation
plt.figure(figsize=(14,5))
plt.xticks(fontname="century", fontsize=14)
plt.yticks(fontname="century", fontsize=14)
ax = data['Acquiror Nation'].value_counts(ascending=False)[0:10].sort_values().plot('barh', color="gray")
for i in ax.patches:
    ax.text(i.get_width()+5, i.get_y(), str(i.get_width()), fontsize=14, fontname="century")
plt.show()
```

Analyze the cross-border M&As by keeping deals in which the target and acquirer nation are different.

```python
international = data[data['Target Nation'] != data['Acquiror Nation']]
international.shape
```

Plot the target and acquirer nations from the cross-border deals.

```python
plt.figure(figsize=(12,5))
plt.xticks(np.arange(0,21,5),fontname="century", fontsize=14)
plt.yticks(fontname="century", fontsize=14)
ax = international['Target Nation'].value_counts(ascending=False)[0:10].sort_values().plot('barh', color="gray")
for i in ax.patches:
    ax.text(i.get_width()+0.5, i.get_y()+0.15, str(i.get_width()), fontsize=14, fontname="century")
plt.show()

plt.figure(figsize=(12,5))
plt.xticks(fontname="century", fontsize=14)
plt.yticks(fontname="century", fontsize=14)
ax = international['Acquiror Nation'].value_counts(ascending=False)[0:10].sort_values().plot('barh',color="gray")
for i in ax.patches:
    ax.text(i.get_width()+0.5, i.get_y()+0.15, str(i.get_width()), fontsize=14, fontname="century")
plt.show()
```
